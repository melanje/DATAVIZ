```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

## Creating a chart

```python
trace0 = go.Scatter(
    x=df.index,
    y=df.NHem,
    name='Northern Hemisphere'
)

trace1 = go.Scatter(
    x=df.index,
    y=df.SHem,
    name='Southern Hemisphere'
)
data = [trace0, trace1]


layout = go.Layout(
    title='Statistics of climate change 1880-2014',
    xaxis=dict(
        title='Year',
        titlefont=dict(
            
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Annual mean temperature ',
        titlefont=dict(
            
            color='#7f7f7f'
        )
    )
)

fig = go.Figure(data=data, layout=layout)

```

```python
iplot(fig, show_link=False)
```
